# LCAO in First Quantisation

## H2 Sigma Bonding

$$s_A = \frac{1}{\pi} \sqrt{\frac{1}{a_0}} e^{-{\frac{|\boldsymbol{r}- R_A|} {a_0}}} \qquad s_B = \frac{1}{\pi} \sqrt{\frac{1}{a_0}} e^{-{\frac{|\boldsymbol{r}-R_B|} {a_0}}}$$


```python
import numpy as np
```

- Ignore constants

```python
def s_orb_a (r):
    return np.exp(-abs(r-1.5))
```

```python
def s_orb_b (r):
    return np.exp(-abs(r+1.5))
```

```python
import math
def sigma_u(r):
    return (math.sqrt(2)**-1)*(s_orb_a(r) + s_orb_b(r))
```

```python
def density(r):
    return (abs(sigma_u(r))**2)
```

```python
r_list = np.linspace(-1,1,100)
```

```python
import pandas as pd

denisty_dict = {r : density(r) for r in r_list}

pd.DataFrame.from_dict(denisty_dict).set_index()
```
